## Elaboração da Análise Fatorial por Componentes Principais

### Teste de Esfericidade de Barlett

Avalia se a aplicação da análise fatorial é adequada ou não.<br>
O teste verifica se os coeficientes de correlação são estatisticamente diferentes de zero.

p-valor < 0,05: rejeita H0

>H0: A Matriz de Correlação é igual à Matriz Identidade.

>H1: A Matriz de Correlação não é igual à Matriz Identidade.

onde,

Matriz Indentidade:
<img src='img\matriz_id.png' style="width: 200px;">


Não queremos que as correlações sejam iguais a zero, pois assim não existiria nenhuma correlação entre as variáveis e consequentemente não seria possível criar fatores.

```python
#https://www.datacamp.com/tutorial/introduction-factor-analysis
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity

# Seleciona os valores métricos
selecao = notasFatorial.iloc[:,1:]

# Calcula p-valor para avaliar se rejeita H0
chi_square_value, p_value=calculate_bartlett_sphericity(selecao)
print()

print(f'ChiSq: {chi_square_value:.4f}, p-value: {p_value:.4f}')
```

### Elaboração da Análise Fatorial por Componentes Principais

```python
#https://factor-analyzer.readthedocs.io/en/latest/factor_analyzer.html#module-factor_analyzer.factor_analyzer
from factor_analyzer import FactorAnalyzer

# Aplicando o metodo fatorial utilziando todas as variáveis da base
fa = FactorAnalyzer(rotation=None, n_factors=len(selecao.columns), method='principal')
fa.fit(selecao)
```

### Eigenvalues ( Autovalores)

O primeiro Autovalor é sempre o maior valor e que gera o fator principal.


Os Autovalores apresentam a proporção da variância que foi compartilhada pelas variáveis originais na formação daquele fator.

63% da variância das variáveis originais já estão representadas no primeiro fator. <br>
25% da variância das variáveis originais estão representadas no segundo fator. <br>
Ambas representam 88% da variância das variáveis originais.

```python
# Obtem os Eigenvalues (autovalores)
ev, v = fa.get_eigenvalues()

# Calcula Variância Compartilhada de cada Fator (fa.get_factor_variance())
"""
var_list=[]
var_cum_list=[]
soma = 0
for autovalor in ev:
    var= autovalor/len(selecao.columns)
    soma = soma + var
    var_list.append(var)
    var_cum_list.append(soma)
"""    

# Cria tabela de autovalores com respectivas variâncias e variâncias acumuladas
tabela_eigen = pd.DataFrame(fa.get_factor_variance()).round(3)
tabela_eigen.columns = [f"Fator {i+1}" for i, v in enumerate(tabela_eigen.columns)]
tabela_eigen.index = ['Autovalor','Variância', 'Variância Acumulada']
tabela_eigen
```

```python
# Resultados dos fatores para as observações do dataset (predict)

predict_fatores= pd.DataFrame(fa.transform(selecao))
predict_fatores.columns =  [f"Fator {i+1}" for i, v in enumerate(predict_fatores.columns)]

predict_fatores
```

```python
# A correlação entre os fatores é zero, os fatores não tem correlação entre eles.

round(predict_fatores.corr(),0)
```

### Scores Fatoriais

Elementos que relacionam a variável em z-score com o fator. <br>
É o peso para cada z-score de cada variável para calcular os fatores.

Fórmula:  Autovalor/ raiz(Autovetor)

No exemplo temos o Fator 1 (PC1) com maior peso em finanças, custos e atuária.<br>
Fator 2 com maior peso em marketing. <br>

O Fator 1 está representando finanças, custos e atuária e o Fator 2 marketing
(Altas correlações (-1 e 1) tendem a formar um único fator, ausência de correlações (0) formam fatores separados)


<img src=img\tabela_scores.png style="width: 300px;">

```python
# Identificando os scores fatoriais

scores = fa.weights_

tabela_scores = pd.DataFrame(scores)
tabela_scores.columns = [f"Fator {i+1}" for i, v in enumerate(tabela_scores.columns)]
tabela_scores.index = selecao.columns
tabela_scores.round(3)
```

### Cargas Fatoriais

São as correlações de pearson entre o fator e a variável original.<br>
Temos a mesma conclusão dos scores fatoriais (pesos).

Fator 1 está representando as matérias de finanças, custos e atuária. <br>
Fator 2 está representando marketing


```python
# Cargas Fatoriais

loadings = pd.DataFrame(fa.loadings_)
loadings.rename(columns = lambda x: 'Fator ' + str(x + 1), inplace=True)
loadings.index = selecao.columns
loadings.round(3)
```

## Comunalidades

São o percentual das variâncias por variável que restaram após a aplicação do critério da raiz latente.

Nesse caso teremos 100% pois estamos considerando todos os fatores o que contempla todas as variâncias das variáveis.

```python
# Calculando comunalidades
communalities = fa.get_communalities()
communalities = pd.DataFrame(communalities, index=selecao.T.index, columns=['communalities'])
communalities.sort_values(by=['communalities'], ascending=False).head(8)
```

# Critério de Kaiser

Seleciona os fatores derivados de autovalores maiores que 1, ou seja, fatores que têm percentual maior de variância das variáveis originais.

```python
#https://www.kaggle.com/code/caesarlupum/factor-analysis-segmentation-analysis/notebook
plt.scatter(range(1,selecao.shape[1]+1),ev)
plt.plot(range(1,selecao.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.axhline(y=1, color='r', linestyle='--')
plt.grid()
plt.show()
```

### Aplicando o Método Utilizando 2 Fatores

```python
# Definindo a quantidade de Fatores > 1
k = len(list(filter(lambda x: x>=1, tabela_eigen.T['Autovalor'])))

# Aplica com 2 fatores seguindo o critério de kaiser

fa2 = FactorAnalyzer(n_factors = k, method = 'principal', rotation = None)
fa2.fit(selecao)
```

Todos os valores abaixo permanecem os mesmos depois do corte:

- Scores;
- Fatores;


Verificar se perdemos Comunalidade com o corte:

Houve uma baixa perda da variância, ainda temos 

- 80% da variância de finanças;
- 87,5% da variância de custos;
- 100% da variância de marketing;
- 84,2% da variância de marketing;

Quanto menor o valor da comunalidade, maior a perda da variância

```python
# Calculando comunalidades
communalities = fa2.get_communalities()
communalities = pd.DataFrame(communalities, index=selecao.T.index, columns=['communalities'])
communalities.round(3)
```

## Criando Ranking com Fatores

Ranking baseado no Fator 1 e Fator 2 ponderando pela variância compartilhada entre os dois fatores.

ou seja,

(Fator 1 * 0,63) + (Fator 2 * 0,25) 

```python
#Adicionando fatores na base:

predict_fatores= pd.DataFrame(fa2.transform(selecao))
predict_fatores.columns =  [f"Fator {i+1}" for i, v in enumerate(predict_fatores.columns)]

notasFatorial['Fator 1'] = predict_fatores['Fator 1']
notasFatorial['Fator 2'] = predict_fatores['Fator 2']
notasFatorial.head()
```

```python
# Criando tabela de Cargas Fatoriais
tabela_cargas = pd.DataFrame(fa2.loadings_)
tabela_cargas.rename(columns = lambda x: 'Fator ' + str(x + 1), inplace=True)
tabela_cargas.index = selecao.columns
tabela_cargas.round(3)

# Criando Tabela de Autovalores
tabela_eigen = pd.DataFrame(fa2.get_factor_variance()).round(3)
tabela_eigen.columns = [f"Fator {i+1}" for i, v in enumerate(tabela_eigen.columns)]
tabela_eigen.index = ['Autovalor','Variância', 'Variância Acumulada']
tabela_eigen
```

```python
tabela_cargas.round(3)
```

```python
# Criando um ranking
notasFatorial['Ranking'] = 0

for index, fator in enumerate(list(tabela_eigen.T.index)):
    variancia = tabela_eigen.T.loc[fator]['Variância']

    notasFatorial['Ranking'] = notasFatorial['Ranking'] + notasFatorial[tabela_eigen.T.index[index]]*variancia

    
notasFatorial = notasFatorial.sort_values(by=['Ranking'], ascending=False)
notasFatorial.head(10)
```

## Gráficos

```python
# Gráfico das cargas fatoriais e suas variâncias nos componentes principais
plt.figure(figsize=(12,8))

tabela_cargas_chart = tabela_cargas.reset_index()

plt.scatter(tabela_cargas_chart['Fator 1'], tabela_cargas_chart['Fator 2'], s=40)

def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x'] + 0.05, point['y'], point['val'])

label_point(x = tabela_cargas_chart['Fator 1'],
            y = tabela_cargas_chart['Fator 2'],
            val = tabela_cargas_chart['index'],
            ax = plt.gca()) 

plt.axhline(y=0, color='black', ls='--')
plt.axvline(x=0, color='black', ls='--')
plt.ylim([-0.5,1.25])
plt.xlim([-0.5,1.25])
plt.title(f"{tabela_eigen.T.shape[0]} componentes principais que explicam {round(tabela_eigen.T['Variância'].sum()*100,2)}% da variância", fontsize=14)
plt.xlabel(f"Fator 1: {round(tabela_eigen.T.iloc[0]['Variância']*100,2)}% de variância explicada", fontsize=14)
plt.ylabel(f"Fator 2: {round(tabela_eigen.T.iloc[1]['Variância']*100,2)}% de variância explicada", fontsize=14)
plt.show()
```

```python
# Gráfico da variância acumulada dos componentes principais

plt.figure(figsize=(12,8))

plt.title(f"{tabela_eigen.shape[0]} componentes principais que explicam {round(tabela_eigen.T['Variância'].sum()*100,2)}% da variância", fontsize=14)
sns.barplot(x=tabela_eigen.T.index, y=tabela_eigen.T['Variância'], data=tabela_eigen.T, color='green')
plt.xlabel("Componentes principais", fontsize=14)
plt.ylabel("Porcentagem de variância explicada", fontsize=14)
plt.show()
```
